```python
import pandas as pd
import numpy as np

dfx=pd.read_excel('ContohData.xlsx')
dfx

x = np.array(dfx.loc[0:20,['Units']])
y = np.array(dfx.loc[0:20,['UnitCost']])

x
```

```python
def normalisasi_minmax(data):
    min_val = np.min(data)
    max_val = np.max(data)
    data = (x-min_val) / (max_val-min_val)
    return data
```

```python
# Data yang telah dinormalisasikan

nor_x =normalisasi_minmax(x)
nor_y =normalisasi_minmax(y)
nor_x
```

```python
def normalisasi_minmax2(data):
    (rows, cols) = data.shape
    mins = np.zeros(shape=(cols), dtype=np.float32)
    maxs = np.zeros(shape=(cols), dtype=np.float32)
    for j in range(cols):
        mins[j] = np.min(data[:,j])
        maxs[j] = np.max(data[:,j])
    result = np.copy(data)
    for i in range(rows):
        for j in range(cols):
            result[i,j] = (data[i,j] - mins[j]) / (maxs[j] - mins[j])
        return (result)
```

```python
normal = normalisasi_minmax2(x)
normal
```
